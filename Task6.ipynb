**WHAT IS DECISION TREE ALGORITHM**: Decision Tree algorithm belongs to the family of supervised learning algorithms. The goal of using a Decision Tree is to create a training model that can use to predict the class or value of the target variable by learning simple decision rules inferred from prior data(training data).

```python
# Importing the required Libraries

from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier, export_graphviz
from sklearn.model_selection import train_test_split
import sklearn.metrics as sm

import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import pydot
from IPython.display import Image
```

# 1 - Loading the Dataset

```python
# Loading Dataset
iris = load_iris()
X=iris.data[:,:] 
y=iris.target
```

# 2 - Exploratory Data Analysis

```python
#Input data 

data=pd.DataFrame(iris['data'],columns=["Petal length","Petal Width","Sepal Length","Sepal Width"])
data['Species']=iris['target']
data['Species']=data['Species'].apply(lambda x: iris['target_names'][x])

data.head()
```

```python
data.shape
```

```python
data.describe()
```

# 3 - Data Visualization by comparing various features

```python
# Input data Visualization
sns.pairplot(data)
```

```python
# Scatter plot of data based on Sepal Length and Width features
sns.FacetGrid(data,hue='Species').map(plt.scatter,'Sepal Length','Sepal Width').add_legend()
plt.show()

# Scatter plot of data based on Petal Length and Width features
sns.FacetGrid(data,hue='Species').map(plt.scatter,'Petal length','Petal Width').add_legend()
plt.show()
```

# 4 - Decision Tree Model Training# 

```python
# Model Training
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=1) 
tree_classifier = DecisionTreeClassifier()
tree_classifier.fit(X_train,y_train)
print("Training Complete.")
y_pred = tree_classifier.predict(X_test)
```

# 5 - Comparing the actual and predicted flower classification

```python
df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred}) 
df
```

# 6 - Visualizing the Trained Model

```python
#Visualizing the trained Decision Tree Classifier taking all 4 features in consideration

export_graphviz(
        tree_classifier,
        out_file="img\desision_tree.dot",
        feature_names=iris.feature_names[:],
        class_names=iris.target_names,
        rounded=True,
        filled=True
)

(graph,) = pydot.graph_from_dot_file('img\desision_tree.dot')
graph.write_png('img\desision_tree.png')

Image(filename='img\desision_tree.png')
```

# 7 - Predicting the output class for some random values of petal and sepal length and width

```python
print("Class Names = ",iris.target_names)

# Estimating class probabilities
print()
print("Estimating Class Probabilities for flower whose petals length width are 4.7cm and 3.2cm and sepal length and width are 1.3cm and 0.2cm. ")
print()
print('Output = ',tree_classifier.predict([[4.7, 3.2, 1.3, 0.2]]))
print()
print("Our model predicts the class as 0, that is, setosa.")
```

# 8 - Calculating the Model accuracy

```python
# Model Accuracy
print("Accuracy:",sm.accuracy_score(y_test, y_pred))
```

The accuracy of this model is 1 since all the 4 features of the iris dataset for creating the decision tree model is taken into account.
